# Building The Network

```python
from keras import layers
from keras import models

model=models.Sequential()
model.add(layers.Conv2D(32, (3,3) , input_shape=(150,150,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3) ,  activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128, (3,3) ,  activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128, (3,3) ,  activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(4,activation='softmax'))
```

```python
model.summary()
```

# **Configuring the model for training**

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy' , 
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc']
              )
```

# **Using ImageDataGenerator to read images from directories**

```python
from keras.preprocessing.image import ImageDataGenerator 

train_datagen=ImageDataGenerator(rescale=1./255)
validation_datagen=ImageDataGenerator(rescale=1./255)

train_generator=train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)

validation_generator=validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)    
```

# **Fitting the Model using a Batch Generator**

```python
history=model.fit_generator(
    train_generator,
    steps_per_epoch=100,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=50
)
```

# **Feature Extraction with Data Augmentation**



```python
from keras import models
from keras import layers

model_da=models.Sequential()
model_da.add(conv_base)
model_da.add(layers.Flatten())
model_da.add(layers.Dense(256,activation='relu'))
model_da.add(layers.Dense(128,activation='relu'))
model_da.add(layers.Dense(1,activation='softmax'))
```

# Training the model end to end with a frozen convolutional base

```python
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers


train_directory=ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_directory=ImageDataGenerator(
    rescale=1./255
)

train_generator1=train_directory.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)

validation_generator1=test_directory.flow_from_directory(
    validation_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)  

model_da.compile(loss='categorical_crossentropy' , 
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc']
              )


history2=model_da.fit_generator(
    train_generator1,
    steps_per_epoch=100,
    epochs=30,
    validation_data=validation_generator1,
    validation_steps=50
)

```

```python
import matplotlib.pyplot as plt
acc=history2.history['acc']
val_acc=history2.history['val_acc']
loss=history2.history['loss']
val_loss=history2.history['val_loss']

epochs=range(1,len(acc)+1)

plt.plot(epochs, acc, '+' , label="Training Acuuracy")
plt.plot(epochs, val_acc , '*', label="Validation_Accuracy")
plt.title("Training And Validation Accuracy")
plt.legend()

plt.figure()

plt.plot(epochs, loss, '+' , label="Training Loss")
plt.plot(epochs, val_loss , '*', label="Validation_Loss")
plt.title("Training And Validation Loss")
plt.legend()

plt.show()
```

```python
test_generator1=test_directory.flow_from_directory(
    
    test_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)

test_loss,test_acc= model_da.evaluate_generator(test_generator1, steps=50)
print('test acc:', test_acc)
```

# **Freezing all layers to a specific one**

```python
conv_base.trainable=True

set_trainable=False
for layer in conv_base.layers:
  if layer.name=='block5_conv1':
    set_trainable=True
  if set_trainable :
    layer.set_trainable=True
  else :
    layer.trainable=False

model_da.compile(loss='categorical_crossentropy' , 
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc']
              )


history2=model_da.fit_generator(
    train_generator1,
    steps_per_epoch=100,
    epochs=30,
    validation_data=validation_generator1,
    validation_steps=50
)        
```
